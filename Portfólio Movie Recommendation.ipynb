```python
filmespivot = avmov.pivot_table(columns = 'IDUSUARIO', index = 'TITULO', values = 'AVALIACAO')
filmespivot.head(4)
```

```python
filmespivot.fillna(0, inplace = True)
filmespivot.head(20)
```

```python
from scipy.sparse import csr_matrix
filmessparse = csr_matrix(filmespivot)
```

```python
type(filmessparse)
```

```python
from sklearn.neighbors import NearestNeighbors
modelo = NearestNeighbors(algorithm = 'brute')
modelo.fit(filmessparse)
```

Aplicando o modelo e fazendo as recomendaÃ§Ãµes

```python
#FILME: 300
distances, sugestions = modelo.kneighbors(filmespivot.filter(items = ['300'], axis=0).values.reshape(1, -1))
for i in range(len(sugestions)):
    print(filmespivot.index[sugestions[i]])
```

```python
#FILME: ALADDIN
distances, sugestions = modelo.kneighbors(filmespivot.filter(items = ['Aladdin'], axis=0).values.reshape(1, -1))
for i in range(len(sugestions)):
    print(filmespivot.index[sugestions[i]]) 
```
