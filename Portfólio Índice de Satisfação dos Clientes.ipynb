```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('dadoshistoricos.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
sat = data['TARGET']
sat.unique()
```

```python
data = data.drop('ID', axis = 1)
data.head()
```

```python
data.shape
```

```python
colunas = data.columns[0:369].tolist()
```

```python
array = data.values
```

```python
type(array)
```

```python
array
```

```python
x = array[:, 0:369]
y = data['TARGET']
```

```python
#normalizando os dados
scaler = MinMaxScaler(feature_range = (0, 1))
rescaledX = scaler.fit_transform(x)
```

```python
norm = pd.DataFrame(rescaledX, columns = colunas)
```

```python
norm.head()
```

```python
norm.shape
```

```python
#PCA para reduzir a dimensionalidade
pca = PCA(n_components = 30) 
data = pca.fit_transform(norm)
```

```python
data.shape
```

```python
data = pd.DataFrame(data)
data.head()
```

```python
#Concatetando a variavel TARGET ao Dataframe
data = pd.concat([data, y], axis=1)
data.head()
```

Modelo

```python
#treinando o modelo
treino = data.sample(frac=0.60, random_state=0)
teste = data.drop(treino.index)
```

```python
xtreino = treino.iloc[:,0:30]
ytreino = treino.iloc[:,30:31]
xteste = teste.iloc[:,0:30]
yteste = teste.iloc[:,30:31]
```

```python
treino.count
```

```python
teste.count
```

```python
#aplicando o método random forest
classificador = RandomForestClassifier(max_depth=5, random_state=0)
classificador.fit(xtreino, ytreino)
ypred = classificador.predict(xteste)
```

```python
#verificando a acurácia do modelo
acc = accuracy_score(yteste,ypred)
acc
```

```python
#matriz de confusão
conf = confusion_matrix(yteste, ypred)
conf
```

```python
from sklearn.metrics import classification_report
print(classification_report(yteste, ypred))
```
