```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
x = h.drop('median_house_value', axis = 1)
y = h['median_house_value']
from sklearn.model_selection import train_test_split
xtreino, xteste, ytreino, yteste = train_test_split(x, y, test_size=0.33, random_state=42)
lrtreino = LinearRegression().fit(xtreino, ytreino)
lrtreino
```

```python
modelres = pd.DataFrame(columns= ['Model', 'R2Score'])
```

```python
ypredtreino = lrtreino.predict(xtreino)
r2scoretreino = metrics.r2_score(ytreino, ypredtreino)
r2scoretreino
```

```python
lrtreino.score(xtreino, ytreino)
```

```python
lrteste = LinearRegression().fit(xteste, yteste)
lrteste.score(xteste, yteste)
```

```python
ypredteste = lrteste.predict(xteste)
r2scoreteste = metrics.r2_score(yteste, ypredteste)
r2scoreteste
```

```python
#XGBoost Regressor
```

```python
from xgboost import XGBRegressor
xgbtreino = XGBRegressor().fit(xtreino, ytreino)
xgbtreino
```

```python
xgbpredtreino = xgbtreino.predict(xtreino)
```

```python
from sklearn.metrics import r2_score
from sklearn import metrics
r2scoretreinoxgb = metrics.r2_score(ytreino, xgbpredtreino)
r2scoretreinoxgb
```

```python
xgbteste = XGBRegressor().fit(xteste, yteste)
xgbteste
```

```python
xgbpredteste = xgbteste.predict(xteste)
```

```python
r2scoretestexgb = metrics.r2_score(yteste, xgbpredteste)
r2scoretestexgb
```

```python
from sklearn.metrics import mean_absolute_error
meantreino = metrics.mean_absolute_error(ytreino, xgbpredtreino)
meantreino
```

```python
meanteste = metrics.mean_absolute_error(yteste, xgbpredteste)
meanteste
```

```python
modelres.loc[len(modelres)] = ['XGBRegression Teste', r2scoretestexgb]
modelres.loc[len(modelres)] = ['XGBRegression Treino', r2scoretreinoxgb]
modelres.loc[len(modelres)] = ['Linear Regression Teste', r2scoreteste]
modelres.loc[len(modelres)] = ['Linear Regression Treino', r2scoretreino]
```
