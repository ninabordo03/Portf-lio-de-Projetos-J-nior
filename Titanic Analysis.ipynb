```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x = t.drop('Survived', axis = 1)
y = t['Survived']
xtreino, xteste, ytreino, yteste = train_test_split(x, y, test_size=0.33, random_state=42)
log = LogisticRegression(random_state=0).fit(xtreino, ytreino)
log
```

```python
from sklearn import svm
svm = svm.SVC().fit(xtreino, ytreino)
svm
```

```python
print(x.shape)
print(y.shape)
y = y.reshape(-1, 1)
print(y.shape)
```

```python
from sklearn.metrics import accuracy_score
xtreinopredlog = log.predict(xtreino)
xtreinopredlog
```

```python
print(xtreino.shape)
print(xteste.shape)
print(ytreino.shape)
print(yteste.shape)
ytreino = ytreino.reshape(-1, 1)
yteste = yteste.reshape(-1, 1)
```

```python
#accuracy score of training data
```

```python
xtreinopredsvm = svm.predict(xtreino)
xtreinopredsvm
```

```python
acc1 = accuracy_score(ytreino, xtreinopredlog)
print(acc1)
acc2 = accuracy_score(ytreino, xtreinopredsvm)
print(acc2)
```

```python
xtestepredlog = log.predict(xteste)
xtestepredlog
```
